```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up plotting style
plt.style.use('default')
sns.set_palette("husl")

# Analyze booking status distribution
print("Booking Status Distribution:")
status_counts = ride['Booking Status'].value_counts()
print(status_counts)
print(f"\nPercentages:")
print((status_counts / len(ride) * 100).round(2))

# Visualize booking status
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
status_counts.plot(kind='bar', color='skyblue')
plt.title('Booking Status Distribution')
plt.xlabel('Status')
plt.ylabel('Count')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
plt.pie(status_counts.values, labels=status_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Booking Status Percentage')

plt.tight_layout()
plt.show()
```

```python
# Date and Time Analysis
ride['Date'] = pd.to_datetime(ride['Date'])
ride['Hour'] = pd.to_datetime(ride['Time']).dt.hour
ride['DayOfWeek'] = ride['Date'].dt.day_name()

# Rides by hour
hourly_rides = ride['Hour'].value_counts().sort_index()
print("=== RIDES BY HOUR ===")
peak_hour = hourly_rides.idxmax()
print(f"Peak hour: {peak_hour}:00 with {hourly_rides[peak_hour]} rides")

# Rides by day of week
daily_rides = ride['DayOfWeek'].value_counts()
print(f"\n=== RIDES BY DAY OF WEEK ===")
print(daily_rides)

# Visualize time patterns
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Hourly distribution
hourly_rides.plot(kind='line', ax=axes[0,0], marker='o', color='blue')
axes[0,0].set_title('Rides by Hour of Day')
axes[0,0].set_xlabel('Hour')
axes[0,0].set_ylabel('Number of Rides')
axes[0,0].grid(True)

# Day of week distribution
daily_rides.plot(kind='bar', ax=axes[0,1], color='green')
axes[0,1].set_title('Rides by Day of Week')
axes[0,1].tick_params(axis='x', rotation=45)

# Success rate by hour
hourly_success = ride.groupby('Hour')['Booking Status'].apply(lambda x: (x == 'Completed').mean() * 100)
hourly_success.plot(kind='line', ax=axes[1,0], marker='s', color='red')
axes[1,0].set_title('Success Rate by Hour')
axes[1,0].set_xlabel('Hour')
axes[1,0].set_ylabel('Success Rate (%)')
axes[1,0].grid(True)

# Monthly trend
monthly_rides = ride.groupby(ride['Date'].dt.to_period('M')).size()
monthly_rides.plot(kind='line', ax=axes[1,1], marker='D', color='purple')
axes[1,1].set_title('Monthly Ride Trends')
axes[1,1].set_xlabel('Month')
axes[1,1].set_ylabel('Number of Rides')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Ratings and Payment Analysis
completed_with_ratings = ride[(ride['Booking Status'] == 'Completed') & 
                           (ride['Driver Ratings'].notna()) & 
                           (ride['Customer Rating'].notna())]

print("=== RATINGS ANALYSIS ===")
print(f"Average Driver Rating: {completed_with_ratings['Driver Ratings'].mean():.2f}")
print(f"Average Customer Rating: {completed_with_ratings['Customer Rating'].mean():.2f}")

# Payment method analysis
payment_counts = ride['Payment Method'].value_counts()
print(f"\n=== PAYMENT METHOD DISTRIBUTION ===")
print(payment_counts)

# Final visualizations
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Ratings distribution
completed_with_ratings['Driver Ratings'].hist(bins=20, ax=axes[0,0], color='skyblue', alpha=0.7, edgecolor='black')
axes[0,0].set_title('Driver Ratings Distribution')
axes[0,0].set_xlabel('Rating')
axes[0,0].set_ylabel('Frequency')

completed_with_ratings['Customer Rating'].hist(bins=20, ax=axes[0,1], color='lightcoral', alpha=0.7, edgecolor='black')
axes[0,1].set_title('Customer Ratings Distribution')
axes[0,1].set_xlabel('Rating')
axes[0,1].set_ylabel('Frequency')

# Payment method distribution
payment_counts.plot(kind='pie', ax=axes[1,0], autopct='%1.1f%%', startangle=90)
axes[1,0].set_title('Payment Method Distribution')
axes[1,0].set_ylabel('')

# Top pickup and drop locations
top_pickups = ride['Pickup Location'].value_counts().head(10)
top_pickups.plot(kind='barh', ax=axes[1,1], color='green')
axes[1,1].set_title('Top 10 Pickup Locations')
axes[1,1].set_xlabel('Number of Rides')

plt.tight_layout()
plt.show()

print(f"\n=== TOP 5 PICKUP LOCATIONS ===")
print(ride['Pickup Location'].value_counts().head())

print(f"\n=== TOP 5 DROP LOCATIONS ===")
print(ride['Drop Location'].value_counts().head())
```

# Pickup-Drop Location Heatmap Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Get top 15 pickup and drop locations
top_pickups_list = ride['Pickup Location'].value_counts().head(15).index.tolist()
top_drops_list = ride['Drop Location'].value_counts().head(15).index.tolist()

# Filter the dataframe for top pickups and top drops
filtered_ride = ride[
    ride['Pickup Location'].isin(top_pickups_list) & 
    ride['Drop Location'].isin(top_drops_list)
]

# Create pickup-drop crosstab for top locations
pickup_drop_matrix = pd.crosstab(
    filtered_ride['Pickup Location'], 
    filtered_ride['Drop Location'], 
    values=filtered_ride['Booking ID'], 
    aggfunc='count'
).fillna(0)

# Reindex to ensure only top 15 locations in correct order
pickup_drop_matrix = pickup_drop_matrix.reindex(index=top_pickups_list, columns=top_drops_list, fill_value=0)

print(f"Heatmap shows ride volume between top {len(top_pickups_list)} pickup and {len(top_drops_list)} drop locations")
print(f"Total rides in heatmap: {pickup_drop_matrix.sum().sum()}")

# Plot heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(
    pickup_drop_matrix, 
    annot=True, 
    fmt='.0f', 
    cmap='YlOrRd', 
    cbar_kws={'label': 'Number of Rides'},
    linewidths=0.5
)
plt.title('Pickup-Drop Location Heatmap\n(Number of Rides between Top Locations)', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Drop Location', fontsize=12, fontweight='bold')
plt.ylabel('Pickup Location', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Show top 10 routes among filtered data
top_routes = filtered_ride.groupby(['Pickup Location', 'Drop Location']).size().reset_index(name='Ride_Count').nlargest(10, 'Ride_Count')
print("\n=== TOP 10 MOST POPULAR ROUTES (Top 15 Locations Only) ===")
for idx, row in top_routes.iterrows():
    print(f"{row['Pickup Location']} → {row['Drop Location']}: {row['Ride_Count']} rides")
```

```python
import folium
import pandas as pd

# Get top routes first
pickup_drop_pairs = ride.groupby(['Pickup Location', 'Drop Location']).size().reset_index(name='Ride_Count')
top_routes = pickup_drop_pairs.nlargest(10, 'Ride_Count')

# Create a map centered on Delhi NCR
delhi_ncr_map = folium.Map(
    location=[28.6139, 77.2090],  # Delhi coordinates
    zoom_start=10,
    tiles='OpenStreetMap'
)
# Sample coordinates for major NCR locations (approximate)
location_coords = {
    'Connaught Place': [28.6315, 77.2167],
    'Khan Market': [28.5984, 77.2319],
    'AIIMS': [28.5672, 77.2100],
    'IGI Airport': [28.5562, 77.1000],
    'Gurgaon': [28.4595, 77.0266],
    'Noida': [28.5355, 77.3910],
    'Faridabad': [28.4089, 77.3178],
    'Dwarka': [28.5921, 77.0460],
    'Saket': [28.5244, 77.2066],
    'Karol Bagh': [28.6519, 77.1909],
    'Nehru Place': [28.5494, 77.2506],
    'Cyber Hub': [28.4951, 77.0890],
    'Greater Noida': [28.4744, 77.5040],
    'Rohini': [28.7041, 77.1025],
    'Vasant Kunj': [28.5200, 77.1591],
    'Lajpat Nagar': [28.5651, 77.2432],
    'Janakpuri': [28.6219, 77.0855],
    'Preet Vihar': [28.6422, 77.2947],
    'Malviya Nagar': [28.5289, 77.2065],
    'Punjabi Bagh': [28.6742, 77.1317],
    'Khandsa': [28.4400, 77.0600],
    'Barakhamba Road': [28.6286, 77.2297],
    'Badarpur': [28.4950, 77.3081],
    'Pragati Maidan': [28.6127,77.2473],
    'Ashram': [28.5706, 77.2514],
    'Basai Dhankot': [28.4580, 77.0680],
    'Lok Kalyan Marg': [28.6139, 77.2090],
    'Narsinghpur': [28.4100, 77.2800],
    'Mehrauli': [28.5125, 77.1857],
    'Madipur': [28.6653, 77.1044],
    'Pataudi Chowk': [28.3250, 76.7800]
}
# Get top pickup and drop locations from our data
top_pickup_locations = ride['Pickup Location'].value_counts().head(15)
top_drop_locations = ride['Drop Location'].value_counts().head(15)

print("Creating interactive map with major NCR locations...")

# Add pickup location markers (red)
for location, count in top_pickup_locations.items():
    if location in location_coords:
        folium.CircleMarker(
            location=location_coords[location],
            radius=min(15, max(5, count/60)),  # Scale marker size by ride count
            popup=f"<b>Pickup: {location}</b><br>Rides: {count}",
            color='red',
            fill=True,
            fillColor='red',
            fillOpacity=0.7,
            weight=2
        ).add_to(delhi_ncr_map)
# Add drop location markers (blue)
for location, count in top_drop_locations.items():
    if location in location_coords:
        folium.CircleMarker(
            location=location_coords[location],
            radius=min(15, max(5, count/60)),  # Scale marker size by ride count
            popup=f"<b>Drop: {location}</b><br>Rides: {count}",
            color='blue',
            fill=True,
            fillColor='blue',
            fillOpacity=0.7,
            weight=2
        ).add_to(delhi_ncr_map)
# Add lines for top routes
print("Adding popular routes to the map...")
for _, route in top_routes.head(5).iterrows():
    pickup_loc = route['Pickup Location']
    drop_loc = route['Drop Location']
    
    if pickup_loc in location_coords and drop_loc in location_coords:
        # Add a line between pickup and drop
        folium.PolyLine(
            locations=[location_coords[pickup_loc], location_coords[drop_loc]],
            weight=max(3, route['Ride_Count']/4),  # Scale line thickness by ride count
            color='green',
            opacity=0.8,
            popup=f"<b>{pickup_loc} → {drop_loc}</b><br>Rides: {route['Ride_Count']}"
        ).add_to(delhi_ncr_map)
# Add legend
legend_html = '''
<div style="position: fixed; 
            bottom: 50px; left: 50px; width: 200px; height: 120px; 
            background-color: white; border:2px solid grey; z-index:9999; 
            font-size:14px; padding: 10px">
<p><b>NCR Uber Map</b></p>
<p><span style="color:red">●</span> Pickup Locations</p>
<p><span style="color:blue">●</span> Drop Locations</p>
<p><span style="color:green">━</span> Popular Routes</p>
<p><small>Size = Ride volume</small></p>
</div>
'''
delhi_ncr_map.get_root().html.add_child(folium.Element(legend_html))
# Display the map
print("Map created successfully!")
print("Red circles: Top pickup locations")
print("Blue circles: Top drop locations") 
print("Green lines: Top 5 most popular routes")
print("Circle size indicates ride volume")

delhi_ncr_map
```
